# filling missing values

```python
dataset['horsepower'].fillna(dataset['horsepower'].mean(),inplace=True)
dataset
```

```python
dataset['horsepower'].isnull().sum()
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dataset.iloc[:,-1]=lb.fit_transform(dataset.iloc[:,-1])
dataset
```

# removing outliers

```python
from scipy import stats
z=np.abs(stats.zscore(dataset))
z
```

```python
np.where(z>3)
```

```python
dataset_no_outliers=dataset[(z<=3).all(axis=1)]
dataset_no_outliers
```

```python
dataset=dataset_no_outliers
dataset
```

```python
dataset.drop(dataset.columns[[-1]],axis=1,inplace=True)
dataset
```

# seperating independent and dependent variables

```python
x=dataset.iloc[:,1:].values
x
```

```python
y=dataset.iloc[:,0].values
y
```

```python
x.shape
```

# one hot encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer([("on",OneHotEncoder(),[6])],remainder="passthrough")
x=ct.fit_transform(x)
x
```

# removing dummy variable

```python
x=x[:,1:]
x
```

```python
x.shape
```

# feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

# train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# evaluation

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100,criterion="mae",random_state=0)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
y_test
```

# checking accuracy

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
